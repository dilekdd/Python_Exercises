Translate the given string to uppercase, replace commas and periods with spaces, and separate the words individually.

```python
text = "The goal is to turn data into information, and information into insight."
text = text.replace('.', ' ' ).replace(',', ' ')
guncel_text = (', '.join([f'"{word}"' for word in text.split()])).upper()
print(f'[{guncel_text}]')
```

Write a function that takes a list as an argument, assigns the odd and even numbers in the list to separate lists, and returns these lists. Multiple ways will be shared. However, Enumarate is prefered one. 

```python
l = [2, 13, 18, 93, 22]
group_l = [[], []]

def tek_cift(l):
    for i, value in enumerate(l):
        if i % 2 == 0:
            group_l[0].append(value)
        else:
            group_l[1].append(value)
    return group_l
tek_cift(l)
```

The list provided below contains the names of students who ranked in the Engineering and Medicine faculties.

The first three students represent the ranking in the engineering, while the last three belong to the medicine.

Use enumerate to print the student rankings specific to each faculty.

```python
ogrenciler = ["Ali","Veli","Ayşe","Talat","Zeynep","Ece"]

sayac = 1

for i, value in enumerate(ogrenciler, start=1):
    if i <= 3:
        print(f"Muhendislik Fakultesi {i}. Ogrenci, {value}")
    if i > 3:
        print(f"Tip Fakultesi {sayac}. Ogrenci, {value}")
        sayac += 1
```

Three lists are provided below. 

The lists contain the course code, credits, and capacity information, respectively.

Use zip to print the course details.

```python
ders_kodu = ["CMP1005","PSY1001","HUK1005","SEN2204"]
kredi = [3,4,2,4]
kontenjan = [30,75,150,25]

ders_bilgileri = zip(ders_kodu, kredi, kontenjan)
for ders, kredi, kontenjan in ders_bilgileri:
    print(f"Kredisi {kredi} olan {ders} kodlu dersin kontenjani {kontenjan} kisidir ")
```

Two sets are provided below.

You are asked to define a function that prints the common elements if the first set includes the second set, or prints the difference of the second set from the first set if it does not.

```python
kume1 = set(["data", "python"])
kume2 = set(["data", "function", "qcut", "lambda", "python", "miuul"])

if kume1.issuperset(kume2):
    print(kume1.intersection(kume2))
else:
    print(kume2.symmetric_difference(kume1))
```
